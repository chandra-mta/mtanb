```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from copy import deepcopy
from Ska.engarchive import fetch as fetch
from Ska.Matplotlib import plot_cxctime
from kadi import events
import tables
from Chandra.Time import DateTime
```

```python
A2B_SWITCH_DATE = '2020:245'
```

```python
def get_start_stop(year):
    start = f'{year}:001:01:00:00.000'
    stop = f'{year + 1}:001:00:00:00.000'
    return start, stop

def plot_hrc_shield_rate(start, stop, ls='solid', color='steelblue'):
    """
    """
    plt.figure(figsize=(15, 10))

    mixed = False
    if stop < A2B_SWITCH_DATE:
        side = 'a'
    else:
        if start > A2B_SWITCH_DATE:
            side = 'b'
        else:
            mixed = True
    
    # Fetch shield telemetry
    if mixed:
        shlda = fetch.Msid(f'2shldart', start=start, stop=A2B_SWITCH_DATE)
        shldb = fetch.Msid(f'2shldbrt', start=A2B_SWITCH_DATE, stop=stop)
        times = np.append(shlda.times, shldb.times)
        vals = np.append(shlda.vals, shldb.vals)
    else:
        shld = fetch.Msid(f'2shld{side}rt', start=start, stop=stop)
        times = shld.times
        vals = shld.vals
        
    plot_cxctime(times, vals // 256, ls=ls,
                 marker='.', color=color, alpha=0.5)
    plt.grid()
    plt.xlabel('Time')
    plt.ylabel('HRC Shield Telemetry, 2SHLDBRT / 256')
    
    plt.ylim(-5, 82)
    plt.title(f'{start[:4]}')
```

```python
years = np.arange(2000, 2022)
```

```python
ii = 0
start, stop = get_start_stop(years[ii])
plot_hrc_shield_rate(start, stop)
```
